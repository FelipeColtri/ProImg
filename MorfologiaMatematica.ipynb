### 1) Utilize o elemento estruturante abaixo:




```python
# Chamando bibliotecas
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Carregando as imagens
img1 = cv2.imread('img1.png')
img2 = cv2.imread('img2.png')
img3 = cv2.imread('img3.png')
img4 = cv2.imread('img4.png')
img5 = cv2.imread('img5.png')
```

```python
# Função para comparar as duas imagens
def show_img(img1, img2, text):
    plt.figure(figsize=(15,15)) 

    plt.subplot(121)
    plt.imshow(img1, cmap="gray", vmin=0, vmax=255)
    plt.title('Imagem Original')
    plt.axis('off')

    plt.subplot(122)
    plt.imshow(img2, cmap="gray", vmin=0, vmax=255)
    plt.title(f'Imagem {text}')
    plt.axis('off')

    plt.show()
```

a) Aplique erosão e dilatação na imagem img1.png disponível no diretório. 

```python
# Definino o kernel para interagir com a imagem
kernel = np.ones((5, 5),np.uint8)
```

```python
img1_erosion = cv2.erode(img1, kernel, iterations=5)
img1_dilate = cv2.dilate(img1, kernel, iterations=5)

show_img(img1, img1_erosion, 'com Erosão')
show_img(img1, img1_dilate, 'com Dilatação')
```

b) Aplique erosão, dilatação, fechamento na imagem img2.png

```python
img2_erosion = cv2.erode(img2, kernel, iterations=1)
img2_dilate = cv2.dilate(img2, kernel, iterations=1)
img2_opening = cv2.morphologyEx(img2, cv2.MORPH_OPEN, kernel)

show_img(img2, img2_erosion, 'com Erosão')
show_img(img2, img2_dilate, 'com Dilatação')
show_img(img2, img2_opening, 'com Abertura')
```

c) Aplique erosão e dilatação na imagem img3.png

```python
img3_erosion = cv2.erode(img3, kernel, iterations=1)
img3_dilate = cv2.dilate(img3, kernel, iterations=1)

show_img(img3, img3_erosion, 'com Erosão')
show_img(img3, img3_dilate, 'com Dilatação')
```

d) Aplique erosão e dilatação usando outro elemento estruturante nestas mesmas imagens. (Defina o kernel usado)

```python
# Definino o novo kernel menor
kernel2 = np.ones((10, 10),np.uint8)

# Modificando as imagens
img1_erosion2 = cv2.erode(img1, kernel2, iterations=5)
img1_dilate2 = cv2.dilate(img1, kernel2, iterations=5)
img2_erosion2 = cv2.erode(img2, kernel2, iterations=1)
img2_dilate2 = cv2.dilate(img2, kernel2, iterations=1)
img3_erosion2 = cv2.erode(img3, kernel2, iterations=1)
img3_dilate2 = cv2.dilate(img3, kernel2, iterations=1)

# Mostrando as imagens
show_img(img1, img1_erosion2, 'com Erosão')
show_img(img1, img1_dilate2, 'com Dilatação')
show_img(img2, img2_erosion2, 'com Erosão')
show_img(img2, img2_dilate2, 'com Dilatação')
show_img(img3, img3_erosion2, 'com Erosão')
show_img(img3, img3_dilate2, 'com Dilatação')
```

### 2) Considere a imagem img4.png

a) Aplique o filtro morfológigo de erosão (3x3). Qual é o efeito na imagem?


```python
# Novo kernel usado 
kernel3 = np.ones((3, 3),np.uint8)
```

```python
img4_erosion = cv2.erode(img4, kernel3, iterations=1)

show_img(img4, img4_erosion, 'com Erosão')
```

b) Aplique o filtro morfológigo de dilatação (3x3). Qual é o efeito na imagem?

```python
img4_dilate = cv2.dilate(img4, kernel3, iterations=1)

show_img(img4, img4_dilate, 'com Dilatação')
```

c) Aplique o filtro morfológigo de abertura (3x3). Qual é o efeito na imagem?

```python
img4_opening = cv2.morphologyEx(img4, cv2.MORPH_OPEN, kernel3)

show_img(img4, img4_opening, 'com Abertura')
```

d) Aplique o filtro morfológigo de fechamento (3x3). Qual é o efeito na imagem?

```python
img4_closing = cv2.morphologyEx(img4, cv2.MORPH_CLOSE, kernel3)

show_img(img4, img4_closing, 'com Fechamento')
```

### 3) Como podemos obter as bordas dos objetos na imagem img5.png?

```python
# Usando bordas de Canny
img5_canny = cv2.Canny(img5, 0, 1)

# Usando o gradiente morfológico
img5_gradient = cv2.morphologyEx(img5, cv2.MORPH_GRADIENT, kernel3)

show_img(img5, img5_canny, 'Com Bordas Usando Canny')
show_img(img5, img5_gradient, 'Com Bordas Usando Gradiente Morfológico')
```
